```python
# imporying the libraries like : numpy , pandas , matplotlib , seaborn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# using pandas library we are reading out the dataset
df = pd.read_csv(r'C:\Users\LENOVO\OneDrive\Desktop\Diwali Sales Data.csv' , encoding ='unicode_escape')
```

```python
# we permanently removing the status and the unnamed1 columns for the dataset
df.drop(['Status' , 'unnamed1'],axis=1 , inplace=True)
```

Exploratory Data Analysis

```python
# Gender
ax = sns.countplot(x='Gender' ,data=df)

for bars in ax.containers:
    ax.bar_label(bars)     
```

```python
# from the above plot we get that there more female customers than males
```

```python
df.groupby(['Gender'] ,as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)    
```

```python
# from the above table we get that female expenses are more than of males
```

```python
sales = df.groupby(['Gender'] ,as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)    
sns.barplot(x='Gender' , y='Amount' ,data=sales)
```

```python
# the above plot shows that females expenses are more than of males
```

```python
df.columns
```

```python
# Age Group
sns.countplot(data = df ,x = 'Age Group' ,hue='Gender')
```

```python
# the above plot shows that age group of (26-35) spends more amount of money specifically females
```

```python
ax = sns.countplot(data = df ,x = 'Age Group' ,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
# marital status
sales = df.groupby(['Age Group','Marital_Status'] , as_index=False)['Amount'].sum().sort_values(by='Amount' ,ascending=False)
sns.barplot(x='Age Group' , y='Amount' ,hue='Marital_Status' ,data=sales)
```

```python
# the above plot shows that which age group of the people according to the marotal status spend more amount of money
```

```python
# States
sales=df.groupby(['State'] , as_index=False)['Orders'].sum().sort_values(by='Orders' ,ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales , x= 'State',y='Orders')
```

```python
# the above plot shows that from uttar pradesh there are maximum number of orders
```

```python
sales=df.groupby(['State'] , as_index=False)['Amount'].sum().sort_values(by='Amount' ,ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales , x= 'State',y='Amount')
```

```python
# the above plot shows that uttar pradesh spends more amount of money
```

```python
# marital status 
ax = sns.countplot(data=df , x='Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# the above plot shows married and unmarried status of people
```

```python
sales=df.groupby(['Marital_Status','Gender'] , as_index=False)['Amount'].sum().sort_values(by='Amount' ,ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales , x= 'Marital_Status',y='Amount', hue='Gender')
```

```python
# the above plot shows the married female spends more amount of money than the married male
```

```python
# occupation 
sns.set(rc={'figure.figsize':(20,15)})
ax = sns.countplot(data=df,x= 'Occupation')
for bar in ax.containers:
    ax.bar_label(bars)
```

```python
# the above plot shows the occupations of the people
```

```python
sales=df.groupby(['Occupation'] , as_index=False)['Amount'].sum().sort_values(by='Amount' ,ascending=False)
sns.set(rc={'figure.figsize':(14,10)})
sns.barplot(data=sales , x= 'Occupation',y='Amount')
```

```python
# the above plot shows that the people from IT sector spends more anount of money
```

```python
df.columns
```

```python
# Product category
sns.set(rc={'figure.figsize':(20,15)})
an = sns.countplot(data=df,x='Product_Category')
for bars in an.containers:
    an.bar_label(bars)
```

```python
# the above plot shows the product categorizes 
```

```python
sales=df.groupby(['Product_Category'] , as_index=False)['Amount'].sum().sort_values(by='Amount' ,ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,15)})
sns.barplot(data=sales , x= 'Product_Category',y='Amount')
```

```python
# the above plot shows that on food , clothing & apparel , electronics & gadgets and footwear people spend more
```

```python
sales=df.groupby(['Product_Category'] , as_index=False)['Amount'].sum().sort_values(by='Amount' ,ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,15)})
sns.barplot(data=sales , x= 'Product_Category',y='Amount')
```

```python
# the above plot shows that food category among the people is more popular 
```

```python
sales=df.groupby(['Product_ID'] , as_index=False)['Orders'].sum().sort_values(by='Orders' ,ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,15)})
sns.barplot(data=sales , x= 'Product_ID',y='Orders')
```
